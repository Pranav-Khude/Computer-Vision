```python
import cv2 
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('images/giraffe.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# print(image)
```

```python
# Flip
# 1 --> flip the image horizontally
# 0 --> flip the image vertically
# -1 --> flip the image both horizontally and vertically
image_horizontal = cv2.flip(image_rgb, 1)
image_vertical = cv2.flip(image_rgb, 0)
image_both = cv2.flip(image_rgb, -1)

# Show the images
plt.subplot(221)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(222)
plt.imshow(image_horizontal)
plt.title('Horizontal Flip')
plt.axis('off')

plt.subplot(223)
plt.imshow(image_vertical)
plt.title('Vertical Flip')
plt.axis('off')
plt.subplot(224)
plt.imshow(image_both)

plt.title('Both Flip')
plt.axis('off')
plt.show()

```

```python
# Rotate
image_rotate_90_clockwise = cv2.rotate(image_rgb, cv2.ROTATE_90_CLOCKWISE)
image_rotate_90_counter_clockwise = cv2.rotate(image_rgb, cv2.ROTATE_90_COUNTERCLOCKWISE)

# Show the images

plt.subplot(221)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(222)
plt.imshow(image_rotate_90_clockwise)
plt.title('Rotate 90 Clockwise')
plt.axis('off')

plt.subplot(223)
plt.imshow(image_rotate_90_counter_clockwise)
plt.title('Rotate 90 Counter Clockwise')
plt.axis('off')
```

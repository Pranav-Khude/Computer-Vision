```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('images/giraffe.jpg')
```

```python
image.shape
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
new_height,new_width = 1300, 1300
resize_image = cv2.resize(image_rgb, (new_width, new_height))
plt.imshow(resize_image)
```

Image Scaling

```python
scale_percent = 50  # percent of original size
s_width = int(image.shape[1] * scale_percent / 100)
s_height = int(image.shape[0] * scale_percent / 100)

scaled_image = cv2.resize(image_rgb, (s_width, s_height),interpolation=cv2.INTER_AREA)
plt.imshow(scaled_image)
```

```python
print('Original image shape:', image.shape)
print('Resized image shape:', resize_image.shape)
print('Scaled image shape:', scaled_image.shape)
```

Interpolation

```python
resize_width = 600
resize_height = 600

# interpolation methods
interpolation_methods = {
    'Nearest': cv2.INTER_NEAREST,
    'Linear': cv2.INTER_LINEAR,
    'Cubic': cv2.INTER_CUBIC,
    'Area': cv2.INTER_AREA,
    'Lanczos4': cv2.INTER_LANCZOS4
}
```

```python
resized_image = {}
for method_name, method in interpolation_methods.items():
    resized_image[method_name] = cv2.resize(image_rgb, (resize_width, resize_height), interpolation=method)
    plt.imshow(resized_image[method_name])
    plt.title(f'{method_name} Interpolation')
    plt.axis('off')
    plt.show()
```
